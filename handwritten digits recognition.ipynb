```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pt
from sklearn.tree import DecisionTreeClassifier
```

```python
data=pd.read_csv(r"C:\Users\CHAITANY\Desktop\train.csv")
print(data) 
```

```python
data=pd.read_csv(r"C:\Users\CHAITANY\Desktop\train.csv").as_matrix()
print(data)
```

```python
clf=DecisionTreeClassifier()
```

```python
#training
xtrain = data[0:21000,1:]
train_label=data[0:21000,0]
clf.fit(xtrain,train_label)
```

```python
#testing_data
xtest=data[21000:,1:]
actual_label=data[21000:,0] #will help in getting the accuracy
```

```python
d=xtest[8] #randomly checking for the 8th element
d.shape=(28,28)# as x_test was not in a matrix form
pt.imshow(255-d,cmap='gray') #black text on white bcakground
pt.show()
```

```python
#so its working fine till here
```

```python
print("just for testing")
print("Give any random index from the dataset:")
n=int(input())
d=xtest[n] #randomly checking for the 8th element
d.shape=(28,28)# as x_test was not in a matrix form
print(clf.predict([xtest[n]]))
pt.imshow(255-d,cmap='gray') #black text on white bcakground
pt.show()
```

```python
# okay! so it matches with the input
```

```python
p = clf.predict(xtest)

count=0
for i in range(0,21000):
    if p[i]==actual_label[i]:
        count+=1
    else:
        0
         
print("Accuracy:",(count/21000)*100,"%")    
```
